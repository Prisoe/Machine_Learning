```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read the data
data = pd.read_csv("student.cleaned.data.csv")
data
```

```python
data.head()
```

```python
# Get the features needed for the regression from the data set
X = data[["traveltime", "studytime", "failures", "famrel", "freetime", "goout", "health"]]
X
```

```python
# Get your target
y = np.where(data['G3'] < 10, 0, 1)
y
```

```python
# training and test the data using only 10% as the test data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.10, shuffle=False)
```

```python
# import the MLPClassifier
from sklearn.neural_network import MLPClassifier
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(10,5,20), random_state=1, max_iter=300)
```

```python
# Fit the train data
mlp.fit(X_train, y_train)
```

```python
# accuracy of a model
mlp.score(X_test, y_test)
```

```python
# Check the class type
mlp.classes_
```

```python
y_test
```

```python
# Get the predicted values for Y
y_pred = mlp.predict(X_test)
y_pred
```

```python
pd.Series(y_pred)
```

```python
# y_test has true target values for test features, y_pred has predicted target values for test features
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
